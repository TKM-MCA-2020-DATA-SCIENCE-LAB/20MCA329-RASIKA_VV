```python
from bs4 import BeautifulSoup
import csv

import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage=requests.get("https://www.flipkart.com/mobiles/smartphones~type/pr?sid=tyy,4io&otracker=undefined_footer_footer")
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
print(soup.prettify())#Prettify view how the tags are nested in the document
```

```python
names=soup.find_all('div',class_='_4rR01T')
```

```python
names
```

```python
Mob_names=[]
for i in range(0,len(names)):
  Mob_names.append(names[i].get_text())
```

```python
Mob_names
```

```python
ratings=soup.find_all('div',class_='_3LWZlK')
```

```python
ratings
```

```python
Mob_rating=[]
for i in range(0,len(names)):
  Mob_rating.append(ratings[i].get_text())
```

```python
Mob_rating
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df['Mobile_names']=Mob_names
```

```python
df['Mobile_Ratings']=Mob_rating
```

```python
df.to_csv('mobile.csv',index=True)
```
